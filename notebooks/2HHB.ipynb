```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
```

```python
from MDAnalysis.coordinates.PDB import PDBReader

u = mda.Universe("../data/2hhb.pdb")

print(f"Number of atoms: {u.atoms.n_atoms}")
print(f"Number of residues: {u.residues.n_residues}")
print(f"Number of segments: {u.segments.n_segments}")
print(f"Available atom names: {u.atoms.names[:10]}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

alpha_carbons = u.select_atoms("name CA")
coordinates = alpha_carbons.positions

ax.scatter(coordinates[:, 0], coordinates[:, 1], coordinates[:, 2], s=20, c='blue')
ax.set_title("Alpha-Carbon Backbone")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
